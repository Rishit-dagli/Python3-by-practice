#### Importing pandas and numpy

```python
import pandas as pd
import numpy as np
```

```python
# Reading the dataset with the pandas data frames
df = pd.read_csv('breast_cancer_data.csv')
```

```python
# Drop these columns as we dont' need it because we've id which doesn't need for featuring also a column unnamed which doesn't have entries.
df = df.drop(['id', 'Unnamed: 32'], axis=1) 
```

```python
# See the first five entries of the dataset
df.head()
```

```python
# Now in the dataset we don't have any null valued column
df.info()
```

```python
# see the dataset now
df
```

### Importing sklearn and some of it's important modules we needed

```python
import sklearn
# svm ==> support vector machine
from sklearn import svm
```

```python
from sklearn.model_selection import train_test_split
# For accuracy score
from sklearn import metrics
```

#### Features

```python
# Drop the dignosis column from the rest as this is our label to predict
feartures = np.array(df.drop(['diagnosis'], 1))
```

#### Label

```python
# Assign diagnosis to the label var
label = np.array(df['diagnosis'])
```

### Split the datset
Splitting the dataset into 80-20%, for training and testing respectively with the help of sklearn

```python
# 20 % of the total data is assigned to the testing purpose
X_train, X_test, y_train, y_test = train_test_split(feartures, label, test_size=0.2)
```

### Training the Model

```python
# Soft margin for this linear kernel is 5
clf = svm.SVC(kernel="linear", C=5)
```

```python
clf.fit(X_train, y_train)
```

#### Evalualtion and Prediction

```python
accuracy = clf.score(X_test, y_test)
accuracy
```

```python
prediction = clf.predict(X_test)
```

```python
prediction
```

```python
acc = metrics.accuracy_score(y_test, prediction)
```
